### numpy 함수 연습 

```python
# 라이브러리 호출 -> numpy를 호출하고 사용할 때는 np로 사용하겠다.
import numpy as np
```

```python
# 리스트를 Numpy 배열로 변환
arr = np.array([1,2,3])
arr
```

```python
# 주어진 범위 내에서 일정한 간격의 값을 가지는 배열 생성
arr = np.arange(0,10,2)
arr
```

```python
# 3행 3열로 0~1 난수로 배열 생성
arr = np.random.rand(3,3)
arr
```

```python
# 1행 9열로 배열 형태 변경
arr = arr.reshape(1,9)
arr
```

```python
# 합계, 평균, 표준편차 구하기
print(np.sum([1,2,3]))
print(np.mean([1,2,3]))
print(np.std([1,2,3]))
```

### 데이터 분석 및 시각화

선생님 Colab Notebooks에 있는 dataset 폴더를 다운받아 파이썬 프로젝트 폴더에 폴더째로 넣기

```python
# 데이터를 불러오기 위한 pandas, 데이터 시각화를 위한 matplotlib, seaborn 호출
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# csv 파일 읽기
df = pd.read_csv('dataset/diabetes.csv')
```

```python
# 불러온 데이터의 위에서 5개 출력
df.head()
```

```python
# 불러온 데이터의 맨 밑에서 5개 출력
df.tail()
```

```python
# 데이터프레임의 기본적인 통계 정보 확인
df.describe()
```

```python
# 데이터프레임의 요약 정보 확인 (인덱스, 컬럼명, 데이터타입 등등..)
df.info()
```

```python
# 데이터프레임의 상관관계 분석 (값이 1이랑 가까울수록 높은 관계를 가지고 있음)
df.corr()
```

```python
# 분석한 상관관계를 seaborn 라이브러리를 사용해 heatmap 작성 (plt를 사용해 보여줌)
sns.heatmap(df.corr(), annot=True)
# 그래프 출력
plt.show()
```

```python
# 히스토그램 생성 (히스토그램은 막대그래프를 이용해 연속형 데이터의 분포를 나타냄)
# figsize는 그래프를 나타낼 도화지의 가로 세로 크기를 의미
df.hist(figsize=(10, 8))
plt.show
```

```python
# 데이터프레임의 널값이 존재하는지 합계를 통해 확인 
df.isnull().sum()
```

```python
# 만약 결측치(널값)이 있다면 fillna를 통해 채워줌 (df.mean = 평균값으로 채운다는 의미)
df.fillna(df.mean(), inplace=True)
df
```

#### 2차시 pdf 61페이지를 참고

```python
# Numpy에 linspace메서드를 통해 (start, stop, num => 0부터 10까지 100개의 간격으로 요소 생성)
x = np.linspace(0,10,100)
# 삼각함수 sin을 사용
y = np.sin(x)
# plot은 선 그래프를 그리는 메서드
plt.plot(x,y)
# x 축 레이블, y축 레이블 설정 
plt.xlabel('x')
plt.ylabel('y')
# 그래프 제목 설정
plt.title('Sine Wave')
# 그래프 출력
plt.show()
```

#### 2차시 pdf 63페이지 seaborn 함수 예시 참고

```python
# tips는 seaborn 라이브러리를 설치하면서 내장돼있는 샘플 데이터셋임
# load_dataset을 통해 tips 데이터셋을 불러옴
tips = sns.load_dataset('tips')
sns.distplot(tips['tip'], kde=True)
plt.show()
```

```python
sns.displot(tips['tip'], kde=True)
plt.show()
```

```python
sns.histplot(tips['tip'], kde=True)
plt.show()
```

```python
# 성별에 따른 팁에 대한 박스플롯
# 박스플롯 부가설명 => https://newsjel.ly/archives/newsjelly-report/14177
sns.boxplot(x='sex' , y='tip', data=tips)
plt.show()
```

```python
# 산점도 (관계를 나타내는 그래프)
# 상관관계 분석에 용이
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```
